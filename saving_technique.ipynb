```python
def show(img,title):
    global num
    num+=1
    ax = plt.subplot(r, c, num)   # ✅ capture subplot as ax
                                  # ax = plt.subplot(...) → gives you the Axes object (the subplot).
                                  # Other time use only plt
    ax.imshow(img)
    ax.set_title(title)
    ax.axis("off")
    # if(num==2):
        # Save first two images as one image
        # plt.savefig('/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/save_subfig.jpg')
    
    # Save only this subplot
    if(num==2):
        extent = ax.get_window_extent().transformed(plt.gcf().dpi_scale_trans.inverted())
        plt.savefig('/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/save_subfig.jpg', bbox_inches=extent, dpi=300)
    plt.grid(False)
```

```python
if __name__=="__main__":
    main()
```
