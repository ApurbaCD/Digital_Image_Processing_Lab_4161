```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def embed_center(big_shape, small_img):
    H, W = big_shape   # big image height, width
    h, w = small_img.shape[:2]

    # Create a large black image
    big_img = np.zeros((H, W, 3), dtype=np.uint8)

    # Compute offsets for centering
    y_offset = (H - h) // 2
    x_offset = (W - w) // 2

    # Copy pixel by pixel (no slicing)
    for i in range(h):
        for j in range(w):
            big_img[y_offset + i, x_offset + j] = small_img[i, j]

    return big_img

# --------------------------
# Example usage
# --------------------------
# Load a small image
small_img = cv2.imread("Image/image5.jpg")
small_img = cv2.resize(small_img, (100, 100))   # resize to 100x100
small_img = cv2.cvtColor(small_img, cv2.COLOR_BGR2RGB)

# Embed into a larger image (300x300)
result = embed_center((300, 300), small_img)
result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
# Display
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(small_img) 
plt.title("Small Image")

plt.subplot(1,2,2), 
plt.imshow(result)
plt.title("Embedded in Large Black Image")
plt.show()
```
