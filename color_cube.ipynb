```python
import matplotlib.pyplot as plt 
import cv2
import numpy as np
#import math

#For display
row = 8
column = 7
fig_num = 1
plt.figure(figsize=(25,18));
def main():
    #For load image
    img_path1 = '/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/image44.jpg'
    img = img_load (img_path1)
    backup_img = img
    h=200
    w=200
    black=np.zeros((h,w,3),dtype=np.uint8)
    white=np.ones((h,w,3),dtype=np.uint8)*255
    display(black,"black")
    intensity(black,[0,1,2])
    display(white,"white")
    intensity(white,[0,1,2])

    fun1(white)
    plt.tight_layout(pad=5.0)
    plt.show()
    plt.close()

def fun1(white):
    h=200
    w=200
    img=np.ones((h,w,3),dtype=np.uint8)
    backup_img=img.copy()
    colors=['red','green','blue']
    channel1=[[0,1],[0,2],[1,2]]
    channel2=[0,1,2]
    # img[:,:,0]=img[:,:,0]*255
    # img[:,:,1]=img[:,:,1]*255 
    # img[:,:,2]=img[:,:,2]*255

    for i in channel2:
        img[:,:,i]=img[:,:,i]*255
        display(img,f"Only {colors[i]}")
        pair=[i]
        img=backup_img.copy()
        intensity(img,pair)
        

    img=backup_img.copy()

    # display(img,"kl")
    for pair in channel1:
        img[:,:,pair[0]]=255
        img[:,:,pair[1]]=255
        display(img,f"{colors[pair[0]]} + {colors[pair[1]]}")
        img=backup_img.copy()
        intensity(img,pair)
        
    
def intensity(img,pair):
    copy_img=img.copy()
    intens=[0,50,100,150,200,255]
    for inten in intens:
        if len(pair)==3:
            img[:,:,pair[0]]=inten
            img[:,:,pair[1]]=inten
            img[:,:,pair[2]]=inten
        elif len(pair)==2 :
            img[:,:,pair[0]]=inten
            img[:,:,pair[1]]=inten
        else:
            img[:,:,pair[0]]=inten
        display(img,f"intencity at {inten}")
        img=copy_img.copy()

    
def img_load(img_path):
    img=cv2.imread(img_path)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    #img=cv2.resize(img,(400,400))
    return img


def display(img,title):
    global fig_num,row,column
    plt.subplot(row,column,fig_num)
    plt.title(title)
    plt.imshow(img)
    fig_num+=1
    

def histrogram(img,title):
    global fig_num,row,column
    height,width=img.shape
    count=np.zeros(256)
    for i in range(height):
        for j in range(width):
            value=img[i][j]
            #print(value)
            count[value]+=1
            
    x=np.arange(256)
    plt.subplot(row,column,fig_num)
    plt.title(title)
    plt.stem(x,count)
    fig_num+=1
    
    
if __name__=='__main__':
    main()
    
    
```
