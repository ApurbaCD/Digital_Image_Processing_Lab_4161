```python
import matplotlib.pyplot as plt
import cv2
def main():
    img_path1 = '/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/image33.jpg'
    img_path2 = '/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/image32.jpg'
    
    img1 = plt.imread(img_path1)
    img2 = plt.imread(img_path2)
    
    c=10
    img3=img1+c    
    # print(img3.shape)
    # print(img3)
    
    img1=cv2.resize(img1, (256, 256))
    img2=cv2.resize(img2, (256, 256))
    
    # add 2 picture
    img4=img1+img2  
    # print(img4.shape)
    # print(img4)
    
    
    plt.figure(figsize=(12,8))
    
    plt.subplot(2,2,1)
    plt.imshow(img1)
    plt.title('Original Image 1')
    
    
    plt.subplot(2,2,2)
    plt.imshow(img2)
    plt.title('Original Image 2')
    
    
    
    plt.subplot(2,2,3)
    plt.imshow(img3)
    plt.title(f'Transfered Image After adding {c} value')
    #plt.grid(True)

    plt.subplot(2,2,4)
    plt.imshow(img4)
    plt.title('Transfered Image After adding image 1 and image 2')
        
    plt.show()

if __name__=='__main__':
    main()
```
