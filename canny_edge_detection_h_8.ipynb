```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

row=2
column=2
num=1

def main():
    img_path1 ='/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/1image.jpg'
    img=cv2.imread(img_path1);
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(20,12))

    show(img,"original image")
    # Convert to grayscale for edge detection
    gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # Apply Canny edge detection
    edges1=cv2.Canny(gray_img,threshold1=100,threshold2=200)
    show(edges1,"Canny Edge Detection 100-200")

    edges2= cv2.Canny(gray_img,threshold1=10,threshold2=20)
    show(edges2,"Canny Edge Detection 10-20")

    plt.tight_layout()
    plt.show()
```

```python
def show(img,title):
    global num
    plt.subplot(row,column,num)
    plt.title(title)
    plt.imshow(img,cmap="gray")
    plt.grid(False)
    num +=1
```

```python

if __name__=='__main__':
    main()
```
