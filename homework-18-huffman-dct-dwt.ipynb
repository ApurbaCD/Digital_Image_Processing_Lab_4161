```python
import numpy as np
import cv2
```

```python
img=cv2.imread('Image/decompressed_image/artificial.pgm')
```

```python
# huffman 
import heapq
from collections import defaultdict

# Build frequency dictionary
def build_frequency(data):
    freq = defaultdict(int)
    for symbol in data:
        freq[symbol] += 1
    return freq

# Build Huffman Tree
def build_huffman_tree(freq):
    heap = [[weight, [symbol, ""]] for symbol, weight in freq.items()]
    heapq.heapify(heap)
    while len(heap) > 1:
        lo = heapq.heappop(heap)
        hi = heapq.heappop(heap)
        for pair in lo[1:]:
            pair[1] = '0' + pair[1]
        for pair in hi[1:]:
            pair[1] = '1' + pair[1]
        heapq.heappush(heap, [lo[0] + hi[0]] + lo[1:] + hi[1:])
    return sorted(heapq.heappop(heap)[1:], key=lambda p: (len(p[-1]), p))

# Example usage
data = "this is an example for huffman encoding"
freq = build_frequency(data)
huffman_codes = build_huffman_tree(freq)
print("Huffman Codes:", huffman_codes)
```

```python
#DCT image compression 
from scipy.fftpack import dct,idct
# Appy 2D dct
dct_image=dct(dct(img.T,norm='ortho').T,norm='ortho')

#quantization
q=np.ones_like(dct_image)*50
dct_quantized=np.round((dct_image)/q)

#recontruction
dct_reconstruct=dct_quantized*q
reconstruct=idct(idct(dct_reconstruct.T,norm='ortho').T,norm='ortho')
cv2.imwrite('Image/decompressed_image/artificial-dct-com.jpg',np.uint8(np.clip(reconstruct,0,255)))
```

```python
#DWT image compression
import pywt

# Apply 2D DWT
coeffs2=pywt.dwt2(img,'haar')
cA,(cH,cV,cD)=coeffs2

# Apply thresholding 
threshold=20
cH=pywt.threshold(cH,threshold,mode='soft')
cV=pywt.threshold(cV,threshold,mode='soft')
cD=pywt.threshold(cD,threshold,mode='soft')

#Reconstruct image

reconstructed=pywt.idwt2((cA,(cH,cV,cD)),'haar')
cv2.imwrite('Image/decompressed_image/artificial-dwt-com.jpg',np.uint8(np.clip(reconstructed,0,255)))
```
