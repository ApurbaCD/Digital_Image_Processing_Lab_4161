```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# histrogram

def main():
    # image1 for black image
    # img_path1 = '/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/image7.jpg' #let consider this is blackish image

    # image16 for white image 
    img_path2 = '/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/image16.jpg'  #let consider this is whitish
    
    # image for normal histogram 
    img_path1 = '/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/image36.jpg'

    img1 = plt.imread(img_path1)
    img2 = plt.imread(img_path2)
    
    #for blackish image
    #img1=cv2.resize(img1,(256,256))
    
    #for whitish image
    img1=cv2.resize(img1,(256,256))
    
    #open cv2 read image in BGR format
    
    #extract every channel image
    img_r=img1[:,:,0]
    img_g=img1[:,:,1]
    img_b=img1[:,:,2]
    
    
    # print(img_r)
    # print(img_g)
    # print(img_b)
    
    plt.figure(figsize=(10,6))
    
    plt.subplot(2,2,1)
    plt.imshow(img1)
    plt.title('Image1')
    
    
    title=['red channel pixel','green channel pixel','blue channel pixel']
    img_channels=[img_r,img_g,img_b]
    colors=['red','green','blue']
    count=np.zeros(256)
    
    for k in range(3):
        for i in range(256):
            for j in range(256):
                channel=img_channels[k]
                count[channel[i][j]]+=1
        x=np.arange(256) #for plotting 
        plt.subplot(2,2,k+1+1)
        plt.grid(True)
        plt.stem(x,count,linefmt=colors[k], markerfmt='.', basefmt=" ")
        plt.title(title[k])
        count=np.zeros(256) 
             
    
    plt.tight_layout()
    plt.show()


  
if __name__=='__main__':
    main()
    
    
  
```
