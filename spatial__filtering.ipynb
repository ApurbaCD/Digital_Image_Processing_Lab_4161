```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

row=4
column=2
num=1
plt.figure(figsize=(15,8))
def main():
    img_path1 ='/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/image24.jpg'
    img=cv2.imread(img_path1,0);
    
    show(img,"original image")

    #Adding noise in the image 
    noise_img=gaussian_noise(img,0,10)

    # Define average filter (3x3)
    k = 3
    average_kernel = np.ones((k, k), np.float32) / (k * k)
    fun(noise_img,average_kernel,"Average kernel")


    prewitt_x=np.array([[-1,0,1],
                      [-2,0,2],
                      [-1,0,1]],dtype=np.float32)
    prewitt_y=np.array([[-1,-2,-1],
                      [0,0,0],
                      [1,2,1]],dtype=np.float32)
    

    sobel_x=np.array([[-1,0,1],
                      [-2,0,2],
                      [-1,0,1]],dtype=np.float32)
    sobel_y=np.array([[-1,-2,-1],
                      [0,0,0],
                      [1,2,1]],dtype=np.float32)
    

    laplacian=np.array([[0,-1,0],
                        [-1,4,-1],
                        [0,-1,0]],dtype=np.float32)
    
    fun(noise_img,prewitt_x,"prewitt_x")
    fun(noise_img,prewitt_y,"prewitt_y")

    fun(noise_img,sobel_x,"sobel_x")
    fun(noise_img,sobel_y,"sobel_y")

    fun(noise_img,laplacian,"laplacian")
    plt.tight_layout()
    plt.show()


def gaussian_noise(img,mean,sigma):

    noise=np.random.normal(mean,sigma,img.shape)
    noise_img = np.clip(img + noise, 0, 255).astype(np.uint8)
    show(noise_img,"Noise Image")
    return noise_img


def fun(img,kernel,title):
    
    filtered_img=cv2.filter2D(img,-1,kernel)
    show(filtered_img,f"after applying {title}")



def show(img,title):
    global num
    plt.subplot(row,column,num)
    plt.title(title)
    plt.imshow(img,cmap="gray")
    plt.grid(False)
    num +=1

    
if __name__=='__main__':
    main()
    
    
  
```
