```python
import matplotlib.pyplot as plt 
import numpy as np
import cv2

fig_num=1
row=4
column=3
plt.figure(figsize=(25,18));

def main():
    img_path='/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/image4.jpg'
    img=cv2.imread(img_path)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    # plt.imshow(img,cmap='gray')
    # plt.show()
    display(img,"Original Image")
    img_copy=img.copy()

    fun(img)

def fun(img):
    h,w=img.shape
    img_copy=img.copy()
    level=[]
    titles=['original','level1','level2','level3','level4','level5','level6','level7','level8']
    for i in range(8):
        tmp_img=np.zeros((h,w))
        for j in range(h):
            for k in range(w):
                value=img[j][k]
                new_value=((value>>i)&1)*pow(2,i)
                tmp_img[j][k]=new_value
        display(tmp_img,titles[i+1])

def display(img,title):
    global fig_num,row,column
    plt.subplot(row,column,fig_num)
    plt.title(title)
    plt.imshow(img,cmap='gray')
    fig_num+=1

def histrogram(img,title):
    global fig_num,row,column
    height,width=img.shape
    count=np.zeros(256)
    for i in range(height):
        for j in range(width):
            value=img[i][j]
            #print(value)
            count[value]+=1
            
    x=np.arange(256)
    plt.subplot(row,column,fig_num)
    plt.title(title)
    plt.stem(x,count)
    fig_num+=1



if __name__=='__main__':
    main()
       
```
