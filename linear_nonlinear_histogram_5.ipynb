```python
import matplotlib.pyplot as plt 
import cv2
import numpy as np
#import math

#For display
row = 3
column = 2
fig_num = 1
plt.figure(figsize=(25,18));
def main():
    #For load image
    img_path1 = '/home/apd/code/CSE_Lab/CSE_4_1_DIP/DIP/Image/image44.jpg'
    img = img_load (img_path1)
    backup_img = img

    display(img,"Original Image")
    histrogram(img,"Original Image Histogram")

    img=linear(img)
    display(img," After applying linear funcion on Image")
    histrogram(img,"Image Histogram after applying linear function")

    img=backup_img

    img=nonlinear(img)
    display(img,"After applying nonlinear funcion on Image")
    histrogram(img," Image Histogram after applying nonlinear function")
    
    plt.tight_layout(pad=5.0)
    plt.show()
    plt.close()


def img_load(img_path):
    img=cv2.imread(img_path)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    #img=cv2.resize(img,(400,400))
    return img


def display(img,title):
    global fig_num,row,column
    plt.subplot(row,column,fig_num)
    plt.title(title)
    plt.imshow(img,cmap='gray')
    fig_num+=1
    

def linear(img):
    #linear function g(x,y)=f(x,y)+2*c
    height,width=img.shape
    c=10
    for i in range(height):
        for j in range(width):
            value=img[i][j]
            value=min(value+(2*c),255)
            #print(value)
            img[i][j]=value
    return img      


def nonlinear(img):
    #nonlinear function g(x,y)=f(x,y)^3
    height,width=img.shape
    c=2
    for i in range(height):
        for j in range(width):
            value=img[i][j]
            value=pow(value,5)
            #print(value)
            img[i][j]=value
    return img     


def histrogram(img,title):
    global fig_num,row,column
    height,width=img.shape
    count=np.zeros(256)
    for i in range(height):
        for j in range(width):
            value=img[i][j]
            #print(value)
            count[value]+=1
            
    x=np.arange(256)
    plt.subplot(row,column,fig_num)
    plt.title(title)
    plt.stem(x,count)
    fig_num+=1
    
    
if __name__=='__main__':
    main()
    
    
```
